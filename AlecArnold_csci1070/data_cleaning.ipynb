```python
import numpy as np
import pandas as pd
import math
import statistics

data_df = pd.read_csv("JEOPARDY_DATA.csv")
data_df.head()
#data_df.tail()
```

```python
data_df.columns = data_df.columns.str.replace(" ","") # Overwriting column names to get rid of the spaces in the names. Replace space with nothing.
data_df.columns
```

```python
data_df.dtypes
```

```python
# m/d/Y is the order the data is currently in before we manipulate it
data_df['AirDate'] = pd.to_datetime(data_df['AirDate'],format='%m/%d/%Y') 
# It is preferable to use year-month-day 
data_df.head()
data_df.dtypes
```

```python
# Assign is used to assign a new column 
data_df = data_df.assign(month = lambda x : x['AirDate'].dt.month) # For every row it is going to provide the month of air date in a new column
data_df.head()
```

```python
# Creating new column
# Axis = 1 is referencing the headers 
data_df['month2'] = data_df.apply(lambda x : x['AirDate'].month, axis = 1) 
data_df.head()
```

```python
#Filters 
#we might want to convert to a list because it can illuminate the structure
# of the data
value_list = data_df['Value'].tolist()
value_list = value_list[0:50]
value_list
```

```python
# What if we want to work with this data but we want to preserve the
# original values?
filtered_list = filter(lambda num: int(num.replace("$","").replace(" ","").replace(",",""))>300,value_list)
```
