```python
import pandas as pd
import numpy as np
import math
import statistics
import matplotlib.pyplot as plt

df = pd.read_csv('CSCI_1070_survey_results.csv')

df.fillna('?', inplace=True)
# Standardize the names of places where everyone is from



df['What state/country were you born in?'] = df['What state/country were you born in?'].str.replace("United States","").str.replace("Missouri", "USA").str.replace("Saint Louis", "").str.replace(" ","").str.replace(",","").str.replace("Illinois","USA").str.replace("NewMexico","USA").str.replace("/U.S","").str.replace("Minnesota","USA").str.replace("Georgia","").str.replace("Benghazi","")

# Standardize the pets in some way
df.rename(columns={"How many pets and of which kind do you have?": "How many pets do you have?"}, inplace=True)
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('None'),'0', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('none'),'0', df["How many pets do you have?"])
df['How many pets do you have?'] = df['How many pets do you have?'].str.replace("2 cats and 1 dog","3")
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('No'),'0', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('two'),'2', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('french bulldog'),'0', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('a Husky'),'1', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('2 miniature'),'2', df["How many pets do you have?"])
df["How many pets do you have?"] = np.where(df["How many pets do you have?"].str.contains('Maincoon'),'1', df["How many pets do you have?"])





# Standardize the favorite food in some way
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('P'),'Pasta', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('pas'),'Pasta', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('teak'),'Steak', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('exican'),'Mexican', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = df['What is your favorite type of food?'].str.replace(".","")
df['What is your favorite type of food?'] = df['What is your favorite type of food?'].str.replace("food","")
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('don'),'?', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = np.where(df['What is your favorite type of food?'].str.contains('Bur'),'Burger', df['What is your favorite type of food?'])
df['What is your favorite type of food?'] = df['What is your favorite type of food?'].str.replace("Food","")
# Standardize the majors
df['What is your major?'] = np.where(df['What is your major?'].str.contains('comp'),'Computer Science', df['What is your major?'])
df['What is your major?'] = np.where(df['What is your major?'].str.contains('Sci'),'Computer Science', df['What is your major?'])
df['What is your major?'] = np.where(df['What is your major?'].str.contains('data'),'Data Science', df['What is your major?'])
df['What is your major?'] = np.where(df['What is your major?'].str.contains('Criminology and Criminal Justice'),'Criminology and Criminal Justice', df['What is your major?'])
df['What is your major?'] = np.where(df['What is your major?'].str.contains('Economics and '),'Criminology and Criminal Justice', df['What is your major?'])
df['What is your major?'] = df['What is your major?'].str.replace(".","")
df['What is your major?'] = np.where(df['What is your major?'].str.contains('Graduated'),'International Business', df['What is your major?'])
    


# Standardize the ages
df['How old are you?'] = df['How old are you?'].str.replace(' years old','')
df['How old are you?'] = np.where(df['How old are you?'].str.contains('19'),'19', df['How old are you?'])



#df.tail(20)




#Create 5 visuals to represent the data

# This visual shows the ages of the students in our class, and how many of each age there are.
plt.hist(df['How old are you?'])
plt.title("Age of Students")
plt.ylabel("# of Students")
plt.xlabel("Age")






```

```python
# This visual shows the amount of pets owned by students, and how many students have each amount of pet.
plt.hist(df['How many pets do you have?'])
plt.title("Number of Students' Pets")
plt.ylabel("# of Students")
plt.xlabel("Pets")
```
