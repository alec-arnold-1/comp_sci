```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from openpyxl import Workbook

og_df = pd.read_excel('merged_data.xlsx')

# Remove unncecessary columns
 
removed_columns = ['Year','Unit','Code Value',]

df = og_df.drop(columns=removed_columns)

# Standardize column names

df.rename(columns={'GDP Per Capita (US$)': 'GDP_per_capita'}, inplace=True)

df.columns = df.columns.str.replace(" ","_")

df.columns = df.columns.str.replace("-","_")

#Turning categorical data to numerical data for region related columns.

categorical_columns = ['Country','Region','Continent']

df = pd.get_dummies(df, columns = categorical_columns,dtype=int)

# Deal with missing values

def handle_nulls(column):
    if column.name in ['Country', 'Region', 'Continent']: # I dont want a mean from a column with binary possibilities
        return column.dropna()
    else:
        return column.fillna(column.mean()) # I want a mean to replace nulls in columns with real numeric values

df = df.apply(handle_nulls)

# Account for multicollinearity concerns


df['Criminal_entities'] = df['Criminal_actors'] + df['Criminal_networks'] # Combine highly correlated columns
df.drop(columns=['Criminal_actors', 'Criminal_networks'], inplace=True)

df['Government_policy'] = df['Law_enforcement'] + df['Prevention'] + df['Judicial_system_and_detention'] + df['National_policies_and_laws'] + df['Victim_and_witness_support']
+ df['Political_leadership_and_governance'] + df['Anti_money_laundering'] + df['Government_transparency_and_accountability'] + df['Economic_regulatory_capacity']

df.drop(columns=['Law_enforcement','Prevention','Judicial_system_and_detention','National_policies_and_laws','Political_leadership_and_governance','Victim_and_witness_support',
                 'Anti_money_laundering','Government_transparency_and_accountability','Economic_regulatory_capacity'],inplace=True)







# data that will go into ML

crime_columns = ['Human_trafficking','Human_smuggling','Arms_trafficking','Flora_crimes','Fauna_crimes','Non_renewable_resource_crimes','Heroin_trade','Cocaine_trade',
                   'Cannabis_trade','Synthetic_drug_trade','Criminal_markets']

ml_df = df.drop(columns=crime_columns)
ml_df2 = ml_df + df['Criminal_markets']
for column in df.columns:
    print(column)
df.to_csv('cleanec_crime3')
```
