```python


import numpy as np
import pandas as pd

#modules for statistics
import math
import statistics


belize_listings_path = "belize_airbnb_listings.csv"
belize_df = pd.read_csv(belize_listings_path)
# belize_df.head()
belize_df.tail()

# Create a new dataframe that takes the 
# original dataset and groups the data by neighborhood 
# and finds the average price, total number of reviews, and max availability_365.


grouped_belize_df = belize_df.groupby("neighbourhood")











# Aggregation Function

#grouped_belize_df[['price']].mean()

#grouped_belize_df[['availability_365']].max()

#grouped_belize_df[['number_of_reviews']].count()

grouped_belize_df[['price','availability_365','number_of_reviews']].agg({'price':'mean','availability_365':'max','number_of_reviews':'count'})


#belize_df.describe()
```
