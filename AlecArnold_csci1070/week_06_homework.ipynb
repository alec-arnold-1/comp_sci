5 pts Clean your dataset to turn categorical values into numerical ones. One-hot encoding is likely the answer, but it depends on the dataset. Your data may have ordinal columns, for example where one-hot encoding is not as appropriate. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split


credit_df = pd.read_csv('Credit_card.csv')

numerical_columns = ['Ind_ID','CHILDREN','Annual_income','Birthday_count','Employed_days','Mobile_phone','Work_Phone','Phone','EMAIL_ID','Family_Members']
categorical_columns = ['GENDER','Car_Owner','Propert_Owner','Type_Income','EDUCATION','Marital_status','Housing_type','Type_Occupation']

encoded_credit_df = credit_df.drop(numerical_columns, axis=1)

encoded_credit_df = pd.get_dummies(credit_df, columns = categorical_columns,dtype=int)

encoded_credit_df.head(20)


# Normalizatization
from sklearn.preprocessing import MinMaxScaler

num_columns_normalize = ['CHILDREN', 'Annual_income', 'Birthday_count', 'Employed_days', 'Mobile_phone', 'Work_Phone', 'Phone', 'EMAIL_ID', 'Family_Members']


num_data = encoded_credit_df[num_columns_normalize]


scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(num_data)


encoded_credit_df[num_columns_normalize] = scaled_data

encoded_credit_df.head(20)



```

(3 pts) Perform univariate linear regression on the dataset. Select your variable to predict. How well did this model perform? Is this a good approach for this dataset? Why or why not? 

```python
# Variable to predict: label

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

label_df = pd.read_csv('Credit_card_label.csv')

merged_df = pd.concat([encoded_credit_df, label_df], axis=1)

merged_df = merged_df.dropna()

X = merged_df.drop('label',axis = 1)
y = merged_df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=32)
regression_model = LinearRegression()
regression_model.fit(X_train, y_train)
y_pred = regression_model.predict(X_test)
y_pred

acc = regression_model.score(X_test, y_test)
print(acc)
# correlation = encoded_credit_df.corr()
# sns.heatmap(correlation, annot=True)

# This model did not perform very well. This is not a good approach for this dataset because the outcome is binary.
# Performance before normalization: 0.07866360772622127
# Performance after normalization: -.1408
```

(8 pts) Perform KNN on this dataset. As part of this, write a function that selects the optimal value of k. How well did this model perform?

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV


def optimal_k_value(X_train, y_train, k_values):
   
    grid = {'n_neighbors': k_values}
    
    knn = KNeighborsClassifier()
    
    search = GridSearchCV(knn, grid)
    search.fit(X_train, y_train)
    
    optimal_k = search.best_params_['n_neighbors']
    
    return optimal_k

X = merged_df.drop('label', axis=1)
y = merged_df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

optimal_k = optimal_k_value(X_train, y_train, [1, 3, 5, 7, 9, 11, 13])

knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train_scaled, y_train)
y_pred = knn.predict(X_test_scaled)


def output_calc(knn, X_test_scaled, y_test):
    accuracy = knn.score(X_test_scaled, y_test)
    print("Accuracy:", accuracy)
    print("Optimal K value:", optimal_k)

accuracy = output_calc(knn, X_test_scaled, y_test)



# accuracy = knn.score(X_test_scaled, y_test)
# print("Accuracy:", accuracy)
# print("Optimal K value:",optimal_k)

# This model performed well.
# Performance before normalization: .8847 accuracy
# Performance after normalization: .8847
```
